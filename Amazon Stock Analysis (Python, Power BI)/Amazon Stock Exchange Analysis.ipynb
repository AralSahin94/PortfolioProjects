# Amazon Stock Exchange Analysis

Dataset can be found in "https://www.kaggle.com/kannan1314/amazon-stock-price-all-time"

-- importing libraries --

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

-- load dataset --

```python
df = pd.read_csv('D:/Downloads/Amazon.csv')
```

-- check info and head --

```python
df.info()
```

```python
df.head()
```

-- Date column is object, needs to be changed into datetime --

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

-- Add column: Open-Close Change Percentage --

```python
df['OpenCloseChange'] = df['Close'] - df['Open']
```

```python
df['ChangePerc'] = ((df['Close'] - df['Open']) / df['Open']) * 100
```

```python
df.head()
```

-- Export to excel as xlsx for Power BI visualization

```python
df.to_excel('AMZNstock.xlsx',index=False)
```
